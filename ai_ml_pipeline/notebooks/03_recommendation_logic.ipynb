```python
import json

# Stand-in curriculum syllabus database dictionary lookup model
SYLLABUS_MAP = {
    "DATA STRUCTURES": ["Arrays & Matrices", "Linked Lists", "Binary Search Trees", "Graph Traversals"],
    "SYSTEM ARCHITECTURE": ["Instruction Pipeline Execution", "Cache Memory Mappings", "ALU Math Operations"],
    "DISCRETE MATHEMATICS": ["Propositional Calculus Logic", "Set Theories Matrices", "Combinatorics Principles"]
}

print("Syllabus lookup matrix compiled.")
```

```python
def formulate_study_vectors(student_records):
    """Processes performance maps and returns clear structural study tasks."""
    study_roadmap = {}
    
    for record in student_records:
        if record["needs_focus"] or record["is_failed"]:
            subject = record["subject_name"].upper()
            core_topics = SYLLABUS_MAP.get(subject, ["Core Chapter Review", "Previous Exam Worksheets"])
            
            study_roadmap[record["subject_name"]] = {
                "urgency_rating": "CRITICAL" if record["is_failed"] else "GPA_BOOSTER",
                "remedial_chapters": core_topics[:3],
                "recommended_weekly_hours": 6.0 if record["is_failed"] else 3.5
            }
            
    return study_roadmap

# Testing the engine on our parsed records array
mock_records = [
    {"subject_name": "Data Structures", "is_failed": True, "needs_focus": True},
    {"subject_name": "System Architecture", "is_failed": False, "needs_focus": True}
]

targeted_plan = formulate_study_vectors(mock_records)
print(json.dumps(targeted_plan, indent=4))
```
